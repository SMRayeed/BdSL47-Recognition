```python
from tensorflow.keras import backend

if backend.image_data_format() == 'channels_first':
    images = images.reshape(images.shape[0], 1, 128, 128)
    input_shape = (1, 128, 128)

else:
    images = images.reshape(images.shape[0], 128, 128, 1)
    input_shape = (128, 128, 1)
    
images = images.astype('float32')
```

```python
test_size = 0.2
test_validation_split_size = 0.5
```

```python
def split_data(images, labels, test_size, test_validation_split_size) :

    train_images, test_validation_images, train_labels, test_validation_labels = train_test_split(images, labels, test_size = test_size)
    test_images, validation_images, test_labels, validation_labels = train_test_split(test_validation_images, test_validation_labels, test_size = test_validation_split_size)
    
    return train_images, train_labels, validation_images, validation_labels, test_images, test_labels
```

```python
train_images, train_labels, validation_images, validation_labels, test_images, test_labels = split_data(images, labels, test_size, test_validation_split_size)
```

```python
print('Number of Training Samples : ', train_images.shape[0])
print('Number of Testing Samples : ', test_images.shape[0])
print('Number of Validation Samples : ', validation_images.shape[0])
```

```python
import numpy as np
from keras.utils import np_utils

labels = 37

train_labels = np_utils.to_categorical(train_labels, labels)
validation_labels = np_utils.to_categorical(validation_labels, labels)
test_labels = np_utils.to_categorical(test_labels, labels)
```

```python
model = Sequential()

model.add(Conv2D(32, (5,5), input_shape = input_shape, activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'))

model.add(Conv2D(64, (5,5), activation='relu'))
model.add(MaxPooling2D(pool_size=(4, 4), strides=(4, 4), padding='same'))
 
model.add(Flatten())

model.add(Dense(296, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(37, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs = 20, batch_size = 30)
```

```python
results = model.evaluate(test_images, test_labels)
testAccuracy = results[1]*100
print('Test accuracy : %.2f' % testAccuracy, "%")
```
